```python
# сигнатуры атак на виндовс (которую - неизвестно), 600+ штук
import pandas as pd
import random
import numpy as np
import json
from pandas.io.json import json_normalize
import openpyxl
with open("B:\Учёба\диплом\сигнатуры\Win32\i1.json") as filename:
    dataset_1 = json.load(filename)
df_1 = pd.DataFrame.from_dict(json_normalize(dataset_1), orient='columns')
df_2 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i2.json")
df_3 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i3.json")
df_4 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i4.json")
df_5 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i5.json")
df_6 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i6.json")
df_7 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i7.json")
df_8 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i8.json")
df_9 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i9.json")
df_10 = pd.read_json("B:\Учёба\диплом\сигнатуры\Win32\i10.json")
```

```python
def preobr(arrs):
    arr = np.array([])
    arr = [np.append(arr, i.split()) for i in arrs]
    return np.unique(np.concatenate([i for i in arr]))
anti_1=preobr(df_1.iloc[0:,-15:].values[0])
```

```python
def new_sigs(arrs):
    new_arrs = np.array(arrs.split())
    return new_arrs[random.choices(np.arange(len(new_arrs)), k=20)]
spise_2_10 = [df_2.iloc[-1:,-1:].values[0][0],df_3.iloc[-1:,-1:].values[0][0]
              ,df_4.iloc[-1:,-1:].values[0][0],df_5.iloc[-1:,-1:].values[0][0]
              ,df_6.iloc[-1:,-1:].values[0][0],df_7.iloc[-1:,-1:].values[0][0],
              df_8.iloc[-1:,-1:].values[0][0],df_9.iloc[-1:,-1:].values[0][0],df_10.iloc[-1:,-1:].values[0][0]]
anti_2_10 = [new_sigs(i) for i in spise_2_10]
anti_2_10
```

```python

win32_sigs = np.unique(np.concatenate([anti_1,anti_2_10]))
df_win32 = pd.DataFrame(win32_sigs, columns = ['Win32'])
with pd.ExcelWriter("B:\Учёба\диплом\сигнатуры\signature.xlsx", mode='a') as writer:
    df_win32.to_excel(writer)
```
