```python
'''В данной программе происходит построение и обучение многослойного персептрона для вирусов.
В качестве обучающего датасета используется весь набор данных, сформированный на прошлом этапе.
Точность модели достигает 77,5%'''
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('B:\Учёба\диплом\datasets\For_persep.csv')
x_train2, y_train2 = df.iloc[:,1:6], df.iloc[:,6:9]
model2 = Sequential()
model2.add(Dense(units=5, activation="relu"))
model2.add(Dense(units=4, activation="relu"))
model2.add(Dense(units=3, activation="sigmoid"))
model2.compile(loss="mse", optimizer = "sgd", metrics=["acc"])
```

```python
history2 = model2.fit(x=x_train2, y=y_train2, epochs=100, validation_split=0.2)
```

```python
plt.plot(history2.history['loss'], label='Training loss')
plt.plot(history2.history['val_loss'], label='Validation loss')
plt.legend()
```

```python
scores = model2.evaluate(x_train2, y_train2)
print("\n%s: %.2f%%" % (model2.metrics_names[1], scores[1]*100))
```

```python
model2.save('perceptron2.h5')
```
